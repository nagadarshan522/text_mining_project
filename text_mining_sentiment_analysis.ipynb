# Text preprocessing

```python
import re
import string
```

```python
#removing punctuations, url's and converting text into lower case
def clean_text1(text):
    text=re.sub(r'@','',text)
    text=text.lower()
    text=re.sub(r'http\S+', '',text)
    text=re.sub(re.escape(string.punctuation),'',text)
    text=re.sub(r'I\x92m','',text)
    #text=re.sub(r'<u+0001f525>','',text)
    return text
round1= lambda x: clean_text1(x)
```

```python
tweet_clean=pd.DataFrame(tweets.Text.apply(round1))
tweet_clean
```

```python
def tokenize(txt):
    tokens=re.split('\W+',txt)
    return tokens

tweet_clean['text_tokenize']= tweet_clean['Text'].apply(lambda x: tokenize(x))
tweet_clean
    
```

```python
#removing stopwords
stopwords= nltk.corpus.stopwords.words('english')
stopwords[0:5]
```

```python
def remove_sw(txt_tokenize):
    tweet_clean=[word for word in txt_tokenize if word not in stopwords]
    return tweet_clean
tweet_clean['txt_nosw']= tweet_clean['text_tokenize'].apply(lambda x: remove_sw(x))
    
tweet_clean
```

```python
nltk.download('wordnet')

wn=nltk.WordNetLemmatizer()
def lemmatization(txt_nosw):
    text=[wn.lemmatize(word) for word in txt_nosw]
    return text

tweet_clean['lemmatized']= tweet_clean.txt_nosw.apply(lambda x: lemmatization(x))
tweet_clean
```

# sentiment analysis

```python
#!pip install textblob
from textblob import TextBlob
```

```python
pol= lambda x :TextBlob(x).sentiment.polarity
sub= lambda x: TextBlob(x).sentiment.subjectivity
```

```python
tweet_clean['polarity']= tweet_clean['Text'].apply(pol)
tweet_clean['subjectivity']=tweet_clean['Text'].apply(sub)
tweet_clean['index']=range(0,len(tweet_clean))
tweet_clean
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.distplot(tweet_clean['polarity'])
```

```python
plt.figure(figsize=(15, 10))
sns.lineplot(y='polarity',x='index',data=tweet_clean)
```

```python
plt.rcParams['figure.figsize']=[16,10]
for index, index in enumerate (tweet_clean.index):
    x= tweet_clean.polarity.loc[index]
    y=tweet_clean.subjectivity.loc[index]
    plt.scatter(x,y, color='blue')
```
